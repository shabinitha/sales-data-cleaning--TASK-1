```python
# Load the Raw Dataset
import pandas as pd

df = pd.read_csv('/content/consumer_electronics_sales_data.csv')
```

```python
#Check for Missing Values
print(df.isnull().sum())   # Count missing values per column
```

```python
# Remove Duplicates
df.drop_duplicates(inplace=True)
```

```python
print(df['customergender'].unique())
```

```python
df['customergender'] = df['customergender'].map({0: 'Male', 1: 'Female'})
```

```python
df['customergender'] = df['customergender'].str.capitalize()
```

```python
print(df.columns)
```

```python
df['ProductPrice'] = pd.to_datetime(df['ProductPrice'])
```

```python
# Rename Columns
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
#fix datatype
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df['customerage'] = pd.to_numeric(df['customerage'], errors='coerce')
df['productprice'] = pd.to_numeric(df['productprice'], errors='coerce')
```

```python
df['productprice'] = pd.to_datetime(df['productprice'], format='%d-%m-%Y', errors='coerce')
print(df.dtypes)
```

```python
#Save the Cleaned Dataset
df.to_csv('consumer_electronics_sales_data.csv', index=False)
```
